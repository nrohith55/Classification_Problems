```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
```

```python
data=pd.read_csv("Company_Data.csv")
```

```python
data.head()
```

```python
data.loc[data['Sales']<=10,'Sales']='High'
```

```python
data.loc[data['Sales']!='High','Sales']='Low'
```

```python
data['Sales']
```

```python
data
```

```python
data=pd.get_dummies(data,columns=['ShelveLoc','Urban','US'],drop_first=True)
```

```python
data
```

```python
data.head()
```

```python
#We have to split the data into train and test later into training and testing data
```

```python
X=data.iloc[:,1:12]
```

```python
Y=data.iloc[:,0]
```

```python
X
```

```python
Y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaled_data=scaler.fit_transform(X)
```

```python
print(scaled_data)
```

```python
X=scaled_data
print(X)
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.33,random_state=42)
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(X_train,Y_train)
```

```python
Y_pred=model.predict(X_test)
```

```python
Y_pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print('classification_report :',classification_report(Y_test,Y_pred))
print('confusion_matrix :',confusion_matrix(Y_test,Y_pred))
print('accuracy_score :',accuracy_score(Y_test,Y_pred))
```
